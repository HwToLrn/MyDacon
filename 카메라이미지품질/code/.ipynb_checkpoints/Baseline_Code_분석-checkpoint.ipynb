```python
# Issues
# 1. numpy type의 npy 확장자로 저장하면 어떤 장점이 있는지 알아본다.
# 2. Image를 분할하여 처리하는 것이 어떤 장점이 있는지 알아본다.
#  - 이미지를 분할하여 sliding window 방식에서 U net의 patch 탐색 방식으로 처리하기 위함이다.
#  - 또한 이미지를 중복하여 탐색하지 않기 때문에 데이터 처리 속도가 향상된다.
#  - 정리하면, image를 분할한 이유는 U net patch 탐색 방식을 위한 것이다.
def split_img(image_files_path: np.ndarray, save_path: str, stride: int=128) -> None:
    # save_path 경로에 파일 생성
    # exist_ok를 False로 설정할 시, 이미 폴더가 존재하면 OsError를 발생시킨다.
    # True로 설정할 시, 이미 폴더가 존재해도 덮어씌우기 형태로 실행된다.
    os.makedirs(f'{save_path}{IMG_SIZE}', exist_ok=True)
    
    idx = 0
    for filepath in tqdm(image_files_path):
        img = cv2.imread(filepath)
        
        for row in range(0, img.shape[0], stride):
            for col in range(0, img.shape[1], stride):
                piece = np.zeros(shape=[IMG_SIZE, IMG_SIZE, 3], dtype=np.uint8)
                temp = img[row:row+IMG_SIZE, col:col+IMG_SIZE, :]
                piece[:temp.shape[0], :temp.shape[1], :] = temp
                np.save(file=f'{save_path}{IMG_SIZE}/{idx}.npy', arr=piece)
                idx += 1
```

```python
def tv_map_func(input_path: List[str], label_path: List[str]):
    input_np = np.load(input_path)
    input_np = input_np.astype(np.float32)/255 # Normalization
    label_np = np.load(label_path)
    label_np = label_np.astype(np.float32)/255 # Normalization
    return input_np, label_np

def augmentation(input_data, label_data):
    input_data, label_data = random_rot(input_data, label_data)
    input_data, label_data = random_flip(input_data, label_data)
    return input_data, label_data

def random_rot(input_data, label_data):
    k = np.random.randint(4) # 0~3 값을 출력한다.
    input_data = np.rot90(input_data, k) # 반 시계 방향으로 90도씩 k번 수행한다.
    label_data = np.rot90(label_data, k)
    return input_data, label_data

def random_flip(input_data, label_data):
    execute = np.random.randint(2) # 0~1 값을 출력한다.
    if execute: # 1 == True면 동작한다.
        input_data = np.fliplr(input_data)
        label_data = np.fliplr(label_data)
    return input_data, label_data
```

```python
train_mini_dataset1 = tf.data.Dataset.from_tensor_slices((train_mini_input_files, train_mini_label_files))
train_mini_dataset2 = train_mini_dataset1.map(map_func=lambda item1, item2: tf.numpy_function(tv_map_func, [item1, item2], [tf.float32, tf.float32]),
                                              num_parallel_calls=tf.data.experimental.AUTOTUNE)
train_mini_dataset3 = train_mini_dataset2.batch(batch_size=BATCH_SIZE)
train_mini_dataset4 = train_mini_dataset3.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
```

# Residual Net 정의

```python
def residual_block(inputs, n_filters: int):
    inputs_norm = tf.keras.layers.BatchNormalization()(inputs)

    # Residual Block
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=3, strides=1, padding='same')(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.LeakyReLU(alpha=ALPHA)(x)
    
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=3, strides=1, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Add()([x, inputs_norm])
    x = tf.keras.layers.LeakyReLU(alpha=ALPHA)(x)
    
    return x
```

# Customize Layers

```python
# 간단하게 Input shape을 설정해서 작업하는 방법도 있다.
# x = tf.keras.Input((256, 256, 3))
# x = tf.keras.layers.Conv2D(filters=3, kernel_size=2, strides=1, padding='same')(x)
```

```python
# Sample code for self-customizing
# you can stack any layers alone in the future
# you can make 'input_layer' that needs a shape of tuple type and 'model' that needs the parameters including inputs and outputs
n_filters = 16
input_layer = tf.keras.Input((256, 256, 3))

convm = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=3, activation=None, padding='same')(input_layer)    
convm = tf.keras.layers.LeakyReLU(alpha=ALPHA)(convm)
convm = tf.keras.layers.BatchNormalization()(convm)

convm = residual_block(inputs=convm, n_filters=n_filters)
convm = residual_block(inputs=convm, n_filters=n_filters)
convm = tf.keras.layers.Dropout(DROPOUT_RATE)(convm)

output_layer = tf.keras.layers.Conv2D(filters=3, kernel_size=3, activation='sigmoid', padding='same')(convm)

model = tf.keras.models.Model(inputs=input_layer, 
                              outputs=output_layer)

model.summary()
```

# Backbone Layers의 구조를 알아본다

```python
# backbone 변수 naming의 의미 : ResNet101V2의 layers 구조를 뼈대로 사용하겠다는 의미
# 다음 Cell에서 backbone.layers[122].output의 구조를 가져와서 연결시켜 layers를 쌓는다. 
# summary method를 통해 구조를 살펴보면 많이 쌓지 않았지만 layers가 깊게 쌓인 것을 볼 수 있다.
backbone = tf.keras.applications.ResNet101V2(weights=WEIGHTS, include_top=False, input_shape=(256, 256, 3))
```

```python
input_layer = backbone.input

conv4 = backbone.layers[122].output
conv4 = tf.keras.layers.LeakyReLU(alpha=ALPHA)(conv4)
pool4 = tf.keras.layers.MaxPooling2D(pool_size=2)(conv4)
pool4 = tf.keras.layers.Dropout(DROPOUT_RATE)(pool4)

convm = tf.keras.layers.Conv2D(filters=n_filters*32, kernel_size=3, activation=None, padding='same')(pool4)    
convm = tf.keras.layers.LeakyReLU(alpha=ALPHA)(convm)
convm = tf.keras.layers.BatchNormalization()(convm)

convm = residual_block(inputs=convm, n_filters=n_filters*32)
convm = residual_block(inputs=convm, n_filters=n_filters*32)
convm = tf.keras.layers.Dropout(DROPOUT_RATE)(convm)

output_layer = tf.keras.layers.Conv2D(filters=3, kernel_size=3,
                                     activation='sigmoid',
                                     padding='same')(convm)

model = tf.keras.models.Model(inputs=input_layer, 
                              outputs=output_layer)

model.summary()
```
